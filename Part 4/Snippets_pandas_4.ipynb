```python
people = {
    "first": ["Corey", 'Jane', 'John'],
    "last": ["Schafer", 'Doe', 'Doe'],
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@gmail.com', 'JohnDoe@gmail.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
df
```

```python
# Video 4 --> Basics of filtering data from dataframes and Series objects
# e.g. If we want to look up our survey data and only look at poople who know Python, we can filter that data out
# or maybe we only want to see results from a specific country, or people that have a specific salary range.
# we can do all that by filtering out data from our Series and df objects 
# Filtering --> the start of every project, separating data we do not want form data we do want
```

```python
# basis for our filters, perform basic comparisons using our df --> get back interesting results
# Let's make a basic comparison and see what the result looks like,
# Let's say we want everyone who has the last name of Dou in our dataframe
df['last'] == 'Doe'
# what we get back is a Series object with boolean values
```

```python
# Let's apply the above filter to our df
filt = (df['last'] == 'Doe') # now filt correspons to the filtered Series object we see above
df[filt] # pass filt (filtered Series object variable) directly to our df

# recall the word filter is a keyword and connot be used directly as a variable 
```

```python
df.loc[filt] # see that this line of code respresent the same as in df[filt]
# .l0c[...] --> used to search/gra especific rows and columns by label 
```

```python
df.loc[filt, 'email']
# first argument --> rows we want, second argument --> columns we want
```

```python
# 'and' and 'or' operators to filter data:
# & for and
# | for or

# get all of the rows where the last name is Doe and the first name is John:
filt = ((df['last'] == 'Doe') & (df['first'] == 'John'))
df.loc[filt]
```

```python
# get the email where the last name is Doe and the first name is John:
df.loc[filt, 'email']
```

```python
# example using the or operator --> use the vertical var character |
# We want a filter where the last name is equal to Schafer or the first name is equal to John
filt = ((df['last'] == 'Schafer') | (df['first'] == 'John'))
df.loc[filt]
# PARENTHESIS MATTER!
```

```python
# we can also get the opposite of a filter
# now, we would want all of the rows where the last name isn't Schafer and the first name isn't John
# --> filt = ((df['last'] != 'Schafer') & (df['first'] != 'John'))
# or we can also do (when defining the original before):
df[-filt]
df.loc[filt]
```
