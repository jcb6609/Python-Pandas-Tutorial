```python
import pandas as pd
```

```python
people = {
    "first": ['Corey', 'Jane', 'John', 'Adam'],
    "last": ['Schafer', 'Doe', 'Doe', 'Doe'],
    "email": ['CoreyMSchafer@gmail.com', 'JaneDoe@gmail.com', 'JohnDoe@gmail.com', 'A@email.com']
}
```

```python
df = pd.DataFrame(people)
df
```

```python
# Sorting data
# Sort by last name (column 'last') --> use the .sort_values() method:
df.sort_values(by='last') # last name (column 'last') with values sorted alphabetically (ascending)
```

```python
# If we wanted to sort these in descending orders, then we can pass-in an argument saying:
df.sort_values(by='last', ascending=False)
```

```python
# maybe you wanna sort on multiple columns --> pass-in a list for these columns.
# Let's say the first column we want to sort on is the 'last' in ascending order,
# and if there are duplicate last names, then we want to sort the 'first' after that.
df.sort_values(by=['last', 'first'], ascending=False)
```

```python
# Sort on multiple columns, but having one in descending order and another one in ascending order
df.sort_values(by=['last', 'first'], ascending=[False, True])
# 'last' in descending (False) and 'first' in ascending (True)
```

```python
# save df and make sort permanent --> use inplace=True:
df.sort_values(by=['last', 'first'], ascending=[False, True], inplace=True)
df
```

```python
# Set back to how it was before --> sorting the index with .sort_index() method:
df.sort_index()
```

```python
# wanting to sort a single column --> sort that single Series obj.
# We want to see sorted last names rather than an entire df sorted, then we can say:
df['last'].sort_values()
```
